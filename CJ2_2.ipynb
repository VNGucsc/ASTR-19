```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,1,100)

def sin_func(val):
    return np.sin(val)

def cos_func(val):
    return np.cos(val)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].plot(x, sin_func(x), color='blue')
axes[1].plot(x, cos_func(x), color='red')

axes[0].set_title('sin(x)')
axes[1].set_title('cos(x)')

plt.savefig("sin_cos_plot.pdf")
```
